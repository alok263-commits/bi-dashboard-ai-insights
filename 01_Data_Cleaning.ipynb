```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Sample - Superstore.csv', encoding='windows-1252')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
# 1. Calculate boundaries for Sales outliers
Q1_sales = df['Sales'].quantile(0.25)
Q3_sales = df['Sales'].quantile(0.75)
IQR_sales = Q3_sales - Q1_sales
lower_sales = Q1_sales - 1.5 * IQR_sales
upper_sales = Q3_sales + 1.5 * IQR_sales

# 2. Calculate boundaries for Profit outliers
Q1_profit = df['Profit'].quantile(0.25)
Q3_profit = df['Profit'].quantile(0.75)
IQR_profit = Q3_profit - Q1_profit
lower_profit = Q1_profit - 1.5 * IQR_profit
upper_profit = Q3_profit + 1.5 * IQR_profit

# 3. Create a new column flagging any row that falls outside these normal boundaries
df['Is_Outlier'] = ((df['Sales'] < lower_sales) | (df['Sales'] > upper_sales) | 
                    (df['Profit'] < lower_profit) | (df['Profit'] > upper_profit))

# 4. See how many outliers we flagged!
print(f"Total outliers flagged: {df['Is_Outlier'].sum()} out of {len(df)} rows")
```
